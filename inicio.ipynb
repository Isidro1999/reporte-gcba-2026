```python
import pandas as pd 
df = pd.read_csv("materiales.csv")
df.head()
len(df)
```

```python
df["Fecha de creación"] = pd.to_datetime(df["Fecha de creación"], errors="coerce")
```

```python
df_2y = df[df["Fecha de creación"] >= "2023-01-01"] ## no tenemos mas materiales pero filtro por las dudas los dos ultimos años
len(df_2y)
```

```python
df_2y["Tags"].head(20)
```

```python
df_2y["Tags"].notna().sum()
```

```python
df_2y["Tags_list"] = df_2y["Tags"].dropna().apply(lambda x: [t.strip() for t in x.split(",")])
```

```python
df_tags = df_2y.dropna(subset=["Tags_list"]).explode("Tags_list")
df_tags.head()
```

```python
df_tags[["Grupo", "Valor"]] = df_tags["Tags_list"].str.split(" - ", n=1, expand=True)
```

```python
df_hierarchy = df_tags.pivot_table(
    index="Material ID",
    columns="Grupo",
    values="Valor",
    aggfunc=lambda x: ', '.join(x.dropna().unique())
).reset_index()
```

```python
df_final = df_2y.merge(df_hierarchy, on="Material ID", how="left")
```

```python
# arreglo la creacion de los ejes nuevos. Los voy a sacar y que tome unicamente la primera etiqueta
# Limpiar eje múltiple
df_final["Eje"] = df_final["Eje"].str.split(",").str[0].str.strip()
df_final["Eje"].value_counts()

```

```python
df_final.to_csv("data/df_final.csv", index=False)
```
