```python
# DIAGNOSTICO TEMPORAL DEL PIPELINE (sin modificar la logica)
import pandas as pd
from pathlib import Path

base_path = Path('.')
materiales_path = base_path / 'materiales.csv'
df_final_path = base_path / 'data/df_final.csv'

print('=== DIAGNOSTICO NOTEBOOK -> APP ===')
print(f'Archivo original: {materiales_path}')
print(f'Archivo final: {df_final_path}')


def detectar_id(df):
    candidatos = ['Material ID', 'material_id', 'id', 'ID']
    for c in candidatos:
        if c in df.columns:
            return c
    return None


# 1) Lectura del CSV original
df = pd.read_csv(materiales_path)
material_id_col = detectar_id(df)

print('\n[1] CSV ORIGINAL')
print(f'- Filas iniciales: {len(df)}')
print(f'- Columnas iniciales: {len(df.columns)}')
if material_id_col:
    print(f'- Columna ID detectada: {material_id_col}')
    print(f'- Materiales unicos: {df[material_id_col].nunique(dropna=True)}')
    dup_rows = df.duplicated(subset=[material_id_col], keep=False).sum()
    print(f'- Filas con ID duplicado: {int(dup_rows)}')
else:
    print('- No se detecto columna ID para medir materiales unicos')


# 2) Parseo de fecha y filtro temporal
df['Fecha de creación'] = pd.to_datetime(df['Fecha de creación'], errors='coerce')
print('\n[2] FECHA')
print(f"- Filas sin fecha valida (NaT): {int(df['Fecha de creación'].isna().sum())}")

df_2y = df[df['Fecha de creación'] >= '2023-01-01'].copy()
print('\n[3] FILTRO >= 2023-01-01')
print(f'- Filas antes: {len(df)}')
print(f'- Filas despues: {len(df_2y)}')
print(f'- Filas perdidas en este filtro: {len(df) - len(df_2y)}')
if material_id_col:
    print(f"- Materiales unicos despues: {df_2y[material_id_col].nunique(dropna=True)}")


# 3) Tags y Tags_list
print('\n[4] TAGS')
print(f"- Filas con Tags nulo: {int(df_2y['Tags'].isna().sum())}")
print(f"- Filas con Tags no nulo: {int(df_2y['Tags'].notna().sum())}")

df_2y['Tags_list'] = df_2y['Tags'].dropna().apply(lambda x: [t.strip() for t in x.split(',')])
print(f"- Filas con Tags_list nulo: {int(df_2y['Tags_list'].isna().sum())}")
print(f"- Filas con Tags_list no nulo: {int(df_2y['Tags_list'].notna().sum())}")


# 4) Dropna + explode
before_dropna_tags = len(df_2y)
df_tags_base = df_2y.dropna(subset=['Tags_list']).copy()
after_dropna_tags = len(df_tags_base)

print('\n[5] DROPNA(tags_list) + EXPLODE')
print(f'- Filas antes dropna(tags_list): {before_dropna_tags}')
print(f'- Filas despues dropna(tags_list): {after_dropna_tags}')
print(f'- Filas perdidas por dropna(tags_list): {before_dropna_tags - after_dropna_tags}')

rows_before_explode = len(df_tags_base)
df_tags = df_tags_base.explode('Tags_list')
rows_after_explode = len(df_tags)
print(f'- Filas antes explode: {rows_before_explode}')
print(f'- Filas despues explode: {rows_after_explode}')
print(f'- Delta por explode (duplicacion): {rows_after_explode - rows_before_explode}')

if material_id_col and material_id_col in df_tags.columns:
    vc = df_tags[material_id_col].value_counts(dropna=True)
    print(f'- Materiales unicos post explode: {df_tags[material_id_col].nunique(dropna=True)}')
    print(f'- Materiales con >1 fila post explode: {int((vc > 1).sum())}')
    print(f'- Max filas de un material post explode: {int(vc.max()) if len(vc) else 0}')


# 5) Extraccion Grupo / Valor y pivot
df_tags[['Grupo', 'Valor']] = df_tags['Tags_list'].str.split(' - ', n=1, expand=True)
print('\n[6] EXTRACCION DE ETIQUETAS')
print(f"- Tags sin separador correcto (' - ') -> Valor nulo: {int(df_tags['Valor'].isna().sum())}")

df_hierarchy = df_tags.pivot_table(
    index='Material ID',
    columns='Grupo',
    values='Valor',
    aggfunc=lambda x: ', '.join(x.dropna().unique())
).reset_index()

print('\n[7] PIVOT (jerarquia)')
print(f'- Filas en df_hierarchy: {len(df_hierarchy)}')
print(f'- Columnas en df_hierarchy: {len(df_hierarchy.columns)}')


# 6) Merge final
rows_before_merge = len(df_2y)
df_final_diag = df_2y.merge(df_hierarchy, on='Material ID', how='left')
rows_after_merge = len(df_final_diag)

print('\n[8] MERGE FINAL')
print(f'- Filas antes merge: {rows_before_merge}')
print(f'- Filas despues merge: {rows_after_merge}')
print(f'- Delta merge: {rows_after_merge - rows_before_merge}')
if material_id_col:
    print(f'- Materiales unicos en df_final (diag): {df_final_diag[material_id_col].nunique(dropna=True)}')

claves = ['Eje', 'Sub Eje', 'Tema', 'Subtema', 'Comuna', 'Mes', 'Año', 'Tipo de material']
print('\n[9] FALTANTES EN CAMPOS CLAVE (df_final en notebook)')
for c in claves:
    if c in df_final_diag.columns:
        print(f'- {c}: {int(df_final_diag[c].isna().sum())} nulos')


# 7) Limpieza final de etiquetas (como celdas 16 y 19)
if 'Eje' in df_final_diag.columns:
    df_final_diag['Eje'] = df_final_diag['Eje'].str.split(',').str[0].str.strip()
if 'Sub Eje' in df_final_diag.columns:
    df_final_diag['Sub Eje'] = df_final_diag['Sub Eje'].str.split(',').str[0].str.strip()
if 'Tema' in df_final_diag.columns:
    df_final_diag['Tema'] = df_final_diag['Tema'].astype(str).str.split(',').str[0].str.strip()
if 'Subtema' in df_final_diag.columns:
    df_final_diag['Subtema'] = df_final_diag['Subtema'].astype(str).str.split(',').str[0].str.strip()

print('\n[10] LIMPIEZA DE ETIQUETAS MULTIPLES')
print(f'- Filas se mantienen en: {len(df_final_diag)}')


# 8) Comparacion con exportado actual
df_check = pd.read_csv(df_final_path)
print('\n[11] COMPARACION CON data/df_final.csv')
print(f'- Filas exportadas: {len(df_check)}')
if material_id_col and material_id_col in df_check.columns:
    print(f'- Materiales unicos exportados: {df_check[material_id_col].nunique(dropna=True)}')

if 'Mes' in df_check.columns:
    bad_mes = df_check['Mes'].isna().sum() + df_check['Mes'].astype(str).str.strip().eq('').sum()
    print(f'- Filas sin Mes (nulo o vacio): {int(bad_mes)}')
if 'Año' in df_check.columns:
    bad_year = df_check['Año'].isna().sum() + df_check['Año'].astype(str).str.strip().eq('').sum()
    print(f'- Filas sin Año (nulo o vacio): {int(bad_year)}')
if 'Comuna' in df_check.columns:
    bad_comuna = pd.to_numeric(df_check['Comuna'], errors='coerce').isna().sum()
    print(f'- Filas sin Comuna valida: {int(bad_comuna)}')

print('\n=== FIN DIAGNOSTICO ===')
```
